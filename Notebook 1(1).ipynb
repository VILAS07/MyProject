```
df['salary_in_lpa'] = pd.to_numeric(df['Salary_LPA'],errors='coerse')
```

```
campus_type_counts = df['Campus_type'].value_counts()
campus_type_counts
```

```
campus_type_counts = df['Employer_Name'].value_counts()
campus_type_counts
```

```
average_salary_per_discipline = df.groupby('Discipline')['Salary_LPA'].mean()
average_salary_per_discipline
```

```
offers_per_employer = df.groupby('Employer_Name').size()
offers_per_employer
```

```
placement_counts = df['Discipline'].value_counts()
plt.bar(placement_counts.index,placement_counts.values)
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.title('Placement Counts')
plt.show()
```
